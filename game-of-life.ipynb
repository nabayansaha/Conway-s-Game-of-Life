## **get the state**

```python
def update_state(matrix):
    padded_matrix = np.pad(matrix, pad_width=1, mode='constant', constant_values=0)
    new_matrix = np.zeros_like(matrix)
    
    rows, cols = matrix.shape
    for row in range(1, rows + 1):
        for col in range(1, cols + 1):
            neighbors_sum = (
                padded_matrix[row-1, col-1] + padded_matrix[row-1, col] + padded_matrix[row-1, col+1] +
                padded_matrix[row, col-1] + padded_matrix[row, col+1] +
                padded_matrix[row+1, col-1] + padded_matrix[row+1, col] + padded_matrix[row+1, col+1]
            )
            
            if padded_matrix[row, col] == 1:
                if neighbors_sum < 2 or neighbors_sum > 3:
                    new_matrix[row-1, col-1] = 0
                else:
                    new_matrix[row-1, col-1] = 1
            else:
                if neighbors_sum == 3:
                    new_matrix[row-1, col-1] = 1
    return new_matrix
```

```python
def look(Matrix):
    hmap = sns.heatmap(Matrix,annot=False)
    plt.show()
    return hmap
```

```python
look(M)
```

## **Running the simulation**

```python
fig, ax = plt.subplots()
def update(frame):
    global M
    ax.clear()
    M = update_state(M)
    sns.heatmap(M, ax=ax, cbar=False, annot=False)
ani = FuncAnimation(fig, update, frames=40, repeat=False)
HTML(ani.to_jshtml())
ani.save('game_of_life.gif', writer='pillow')
```
